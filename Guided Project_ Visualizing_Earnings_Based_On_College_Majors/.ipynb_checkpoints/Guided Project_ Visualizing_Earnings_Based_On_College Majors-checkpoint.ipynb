## Visualizing Earnings based on college majors

In this notebook we are goint to do a visual analysis using Viasualization tools like `pyplot` and `matplotlib`. We'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by [American Community Survey](https://www.census.gov/programs-surveys/acs/), which conducts surveys and aggregates the data. FiveThirtyEight cleaned the dataset and released it on their [GitHub](https://github.com/fivethirtyeight/data/tree/master/college-majors) repo.

+ **Do students in more popular majors make more money?**
    + Well taking a look at the Data, we can see that the higher the sample size, the higher the number of students where enrolled in that major. Looking at the first plot, we can see that the data is very disperse around small sample sizes. This is an indicator, that in those majors that are not very popular some make a lot of money and others make not very much. But when the sample size increases we see that the salaries are not very high, but seem to be stable around the same average value. One thing to notice (this is only a assumption), is that the more people with a common majors goes into the market, the chances of them breaking out in the curve is very rare. 
     + Taking a look a the second plot regarting the unemployment Rate, we see here again a similar behaviour like the first plot with the salaries. Here not very popular majors can be secure jobs as well as a high rate of unemployed professionals. But the bigger the sample size, the unemployment rate stays rather constant.
+ **Do students that majored in subjects that were majority female make more money?**
    + If we take a look a the plots 5 and 6, we can see the Men and Women population against the Median salary. The type of the data distribution seem to be very similar around the 0 to 25K mark. But things start to change in majors where the majority where Men. In Majors where the majority where women, the salaries tend to stabilize and do not go over the 50K mark a year, with a lot of values in the low 30K. By men this curve starts do decrease later, and there is more Men earning more, regarding the same population size. Here the max value of Women  is 175K while men goes over the 300K mark. So just by analyzing both plots to the same popuilation size, we can observe that Men earn more money than Women.
    + This relationship is clearer in plot 7, where the Share Women is plotted agains the Median income. Here the correlation is negative. Majors where the majority where Women tend to earn less money.
    
+ **Is there any link between the number of full-time employees and median salary?**
    + Jus by observing the curve the distribution looks very similar to the Men and Women Plots. But in my opinio it is difficult to see if there is any correlation between more people working full time and the return in Salary. This depends mostly on the individual and how it developes in the market. There is in my opinion no generalization for this kind of relationship possible.

### hexagonal bin plot to visualize the columns that had dense scatter plots from earlier in the project.

```python
fig = plt.figure(figsize=(20,5))

ycols = ["Median", "Unemployment_rate", "Total"]
xcols = ["Sample_size", "Sample_size", "Sample_size"]

for i in range(len(ycols)):
    ax = fig.add_subplot(1,len(ycols),i+1)
    recent_grads.plot.hexbin(x= xcols[i], y= ycols[i], ax = ax)
```
