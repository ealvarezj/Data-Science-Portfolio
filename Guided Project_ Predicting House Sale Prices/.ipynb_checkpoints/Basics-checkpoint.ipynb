Let's start by setting up a pipeline of functions that will let us quickly iterate on different models.





```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import KFold
```

## Utility functions
These functions summarize the whole wokflow described in the diagram above. How the function where created and other criteria can be found after the functions declaration

```python
def transform_features(df, drop_thresh = 0.25, variance_tresh = 0.95):
    # drop cols with more than 25% of missing values
    drop_cols = df.dropna(axis = 1, thresh = (df.shape[0]*drop_thresh)).copy()
    
    # fill nan with mean for numerical columns
    numeric = drop_cols.select_dtypes(include=['float64', 'int64', 'bool']).columns
    drop_cols[numeric] = drop_cols[numeric].fillna(drop_cols[numeric].mean())
    
    # drop columns containing information about the sale
    sale_cols = [col for col in list(drop_cols.columns) if ('Sale' in col) | ('Sold' in col)]
    if 'SalePrice' in sale_cols: sale_cols.remove('SalePrice')  
    drop_cols = drop_cols.drop(sale_cols, axis=1)
    
    # Create a new feature from the years
    drop_cols['Year_until_remod'] = drop_cols['Year Remod/Add'] - drop_cols['Year Built']
  
    return df, drop_cols
```

```python
def select_features(df, variance_tresh = 0.95):
    
    # transform some numerical features --> see explanation in analysis
    for col in ['Overall Qual', 'Overall Cond', 'MS SubClass']:
        df[col] = df[col].astype('object')
    
    #   remove low variance columns
    cat_cols = list(df.select_dtypes(include=['object']).columns)
    low_variance_cols = list()
    for col in cat_cols:
        value_counts = df[col].value_counts(normalize = True)
        for value, index in zip(value_counts, value_counts.index):
            if value > variance_tresh:
                low_variance_cols.append(col)
    df = df.drop(low_variance_cols, axis=1)
    
     # convert cols to categorical since all the object cols are categorical
    for col in df.select_dtypes(include=['object']).columns:
        df[col] = df[col].astype('category')
        

    # append the categorical features to the dataframe
    categorical_cols = list(df.select_dtypes(include=['category']).columns)
    categorical_array = pd.get_dummies(df[categorical_cols])
    
    df.drop(categorical_cols, inplace = True, axis=1)
    
    df = pd.concat([df, categorical_array], axis=1)
    
    return df
    
```

```python
def train_and_test(df, k = 0):
    # Define the feature columns by droppping the prediction col
    x_features = list(df.drop(['SalePrice'], axis=1).columns)
    # generate a LinerRegressor
    model = LinearRegression()
    
    # For K = 0 perform Holdout validation
    # For K = 1 since sklearn starts at K = 2 we do Holdout, test and train on both folds --> Take Average RMSE
    # For K >1 use Kfolds index genaratos and loop --> Take average RMSE
    
    if k == 0:
        half = int((df.shape[0])/2)
        train = df.iloc[:half, :]
        test = df.iloc[half:, :]
        
        X = train[x_features]
        y = train.SalePrice
        model.fit(X,y)
        predictions = model.predict(test[x_features])

        mse = mean_squared_error(test.SalePrice, predictions)
        rmse = np.sqrt(mse)
        return rmse, k
        
    elif k == 1:
        half = int((df.shape[0])/2)
        fold_one = df.iloc[:half, :]
        fold_two = df.iloc[half:, :]
        rmse_lst = list()
        
        for train, test in zip([fold_one, fold_two], [fold_two, fold_one]):
            X = train[x_features]
            y = train.SalePrice
            model.fit(X,y)
            predictions = model.predict(test[x_features])
            
            mse = mean_squared_error(test.SalePrice, predictions)
            rmse = np.sqrt(mse)
            rmse_lst.append(rmse)
            
        rmse = np.mean(rmse_lst)
        return rmse, rmse_lst, k
    
    else:
        kf = KFold(n_splits=k, shuffle=True, random_state=1)
        rmse_lst = list()
        for train_index, test_index in kf.split(df):
            train = df.iloc[train_index]
            test  = df.iloc[test_index]
            X = train[x_features]
            y = train.SalePrice
            model.fit(X,y)
            predictions = model.predict(test[x_features])
            
            mse = mean_squared_error(test.SalePrice, predictions)
            rmse = np.sqrt(mse)
            rmse_lst.append(rmse)
        return rmse, rmse_lst, k 
    
```

### Dummy variables and variance based dimensionality reduction
Before we can continue most of the columns in the dataset are from the `object` type. This means they are strings , and from the dataset [documentation](https://s3.amazonaws.com/dq-content/307/data_description.txt) most of them are categorical. With the exeptio of some columns: `Overall Qual` and `Overall Cond`. These are in a scale from 1 to 10 and their meaning code is explained as `Very Poor` to `Very Excelent`. Same goes to the column `MS SubClass`. Here the code from 20 to 190 describes the dwelling involved in the sale. We might want to transforme these features also to categorical.

Let's start by lokking at the sum of unique values for each `object` column (Note the 3 columns mentioned above are not here. These are going to be cleaned in the pipeline)

```python
ames_df.select_dtypes(include=['object']).describe().loc['unique'].sort_values(ascending = False)
```

The Neighborhood column contains the most unique values followed by the Exterior 1 and 2. Since these may be converted to categorical, we need tpo take into account that the dimensionality of the dataset will increase. We can calculate this in forehand by adding all the unique values in all the columns together

```python
ames_df.select_dtypes(include=['object']).describe().loc['unique'].sum()
```

This would leave us with 268 additional columns (plus some more from the 3 columns above).

Now that we have the columns that we want to convert to categorical, we can reduce the dimension  of the dataset, by looking at the variability (variance) of the unique values in each colum. Since these columns do not contain numerical data, it is impossible to calculate the variance. Therefore we can iterate over each column and see the distribution of unique values on each. Then append those with values higher than 95\% to a list and remove them from the analysis. Since these do not add any value to the model.

```python
cat_cols = list(ames_df.select_dtypes(include=['object']).columns)

low_variance_cols = dict()

for col in cat_cols:
    value_counts = ames_df[col].value_counts(normalize = True)
    for value, index in zip(value_counts, value_counts.index):
        if value > 0.95:
            low_variance_cols[col] = [value, index]
low_variance_cols
```

After going over all cols, we can see there are some columns where the 'variance' is very low, since one unique value makes more than 95\% of the data. These columns could be removed from the categorical transformation.

```python
original_df, new_df = transform_features(ames_df)
selected = select_features(new_df)
```

```python
rmses_plot = list()
for k in range(5):
    rmse = train_and_test(selected, k = k)[0]
    rmses_plot.append(rmse)
    
```

```python
sns.lineplot(x = range(5), y = rmses_plot)
plt.title('RMSE over k')
plt.xlabel('k')
plt.ylabel('RMSE')
plt.show()
```
